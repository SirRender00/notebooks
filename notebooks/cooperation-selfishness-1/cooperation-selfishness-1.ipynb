```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
out = Path("./assets")
out.mkdir(exist_ok=True)
```

```python
def u(x):
    return np.log(x)
```

```python
X = np.linspace(1, 20)
plt.plot(X, u(X))

plt.xlabel("x")
plt.ylabel("log(x)")

plt.savefig(out / "log.png")
plt.show()
```

```python
def u(x):
    return np.log(x) + np.log(10-x)
```

```python
X = np.linspace(0, 10)
plt.plot(X, u(X))

plt.xlabel("x")
plt.ylabel("u(x)")

plt.savefig(out / "multivar-utility.png")
plt.show()
```

```python
def u(x, alpha=0.75, beta=0.25):
    return alpha * np.log(x) + beta * np.log(10-x)
```

```python
X = np.linspace(0, 10)
plt.plot(X, u(X))

plt.xlabel("x")
plt.ylabel("u(x)")

plt.savefig(out / "multivar-utility-pref.png")
plt.show()
```

```python
def u(x, alpha=0.25, beta=0.75):
    return alpha * np.log(50 + x) + beta * np.log(100-x)
```

```python
X = np.linspace(0, 100)
plt.plot(X, u(X))

plt.xlabel("x")
plt.ylabel("u(x)")

plt.savefig(out / "bob-fixed-alice.png")
plt.show()
```
